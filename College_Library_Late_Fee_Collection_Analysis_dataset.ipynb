```python
cols_to_keep = ['Book_ID', 'Department', 'Issue_Date', 'Due_Date',
                'Late_Days', 'Return_Date', 'Fine_Amount']
```

```python
df2 = data[cols_to_keep]
```

```python
df2.head()
```

```python
df2.groupby('Department')["Fine_Amount"].count().sort_index()
```

```python
df2['Fine_Amount'].describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(x="Department", data=df2, kde=True, color="brown")
plt.title("Department vs Fine Amount")
plt.xlabel("Department")
plt.ylabel("Fine Amount")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(x="Late_Days", data=df2, kde=True, color="brown")
plt.title("Late Days vs Fine Amount")
plt.xlabel("Late Days")
plt.ylabel("Fine Amount")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Department", y="Fine_Amount", data=df2, color="brown")
plt.title("department vs Fine Amount")
plt.xlabel("departments")
plt.ylabel("Fine Amount")
plt.show()
```
